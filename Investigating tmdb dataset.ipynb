# Cleaning the data

```python
# searching for the duplicate data and eliminating it
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
#eliminating some of the columns that will not be used in this analysis
df.drop(columns = ['homepage', 'production_companies', 'tagline', 'keywords', 'imdb_id'])
```

```python
# Removing null values from genre for better analysis
df = df[df['genres'].isnull()==False]
```

```python
# Checking the null values and filling them
df.isnull().sum()
df.fillna(0)
```

```python
# Changing format of release_date to datetime format
df['release_date'] = pd.to_datetime(df['release_date'])
```

```python
# Adding a column that shows the net profit
df['profit'] = df['revenue_adj']-df['budget_adj']
df.head(3)
```

# Analyzing the dataset

Question: Which director produced the most movies

```python
df['director'].value_counts().head()
```

From the above it is clear that Woody Allen has produced the most movies. He has produced a total of 45 movies and is folled by Clint Eastwood who has produced 34 movies.

Question: Which actor has cast most movies

```python
df['cast'].value_counts().head()
```

It can be clearly deduced that Louis.C.K has cast the most movies. He is closely followed behind by William Shatner, Leornard Nimoy, DeFrost Keelley, James Doohan, and George Takei, who have each been cast in a total of 5 movies

Question: What is the yearly release of movies

```python
# count of movies released per year
yearly_release = df.groupby('release_year').count()['id']
yearly_release.head()
```

```python
yearly_release.tail()
```

```python
# plot movies released by year
plt.figure(figsize=(10, 6))
plt.plot(yearly_release)
# title and labels
plt.title('Number of movies relased per year')
plt.xlabel('release_year')
plt.ylabel('Number of Movies')
```

There is a steady increase in the number of movies being released as seen from the graph above. In 2014 there was the highest number movies produced. 699 movies were produced in that year. After 2014 the number of movies produced is then seen to be declining.

Question: What is the correlation between budget and revenue

```python
plt.figure(figsize = (12, 8))
scatterplot = plt.scatter(df['revenue_adj'], df['budget_adj'], alpha = 0.6)
plt.title('Correlation between budget and revenue', fontsize = 15)
plt.xlabel('revenue', fontsize = 14)
plt.ylabel('budget', fontsize = 14)
plt.show()
```

From the scatterplot above, it can be clearly seen that there is a positive correlation between the adjusted budget and the adjusted revenue. We use the adjusted values because it gives the actual values that were referenced. We can also say that for the movies with proper budgeting gave a good revenue.

Question: How popularity influences and is influenced by various factors in the dataset.

```python
# plot of vote count vs popularity
plt.figure(figsize=(12, 8))
plt.scatter( df['vote_count'], df['popularity'])
plt.title('Vote count vs Popularity', fontsize = 15)
plt.xlabel('vote count', fontsize = 14)
plt.ylabel('Popularity rating', fontsize = 14)
```

There is a positive correlation between vote count and popularity. This is to say that movies with a higher vote count are more popular compared to movies with a low vote count.

```python
#plot of popularity vs revenue
plt.figure(figsize=(12, 8))
plt.scatter(df['popularity'], df['revenue_adj'])
plt.title('Popularity vs Revenue',fontsize = 15)
plt.xlabel('Popularity rating', fontsize = 14)
plt.ylabel('Revenue', fontsize = 14)
```

The higher the popularity rating, the more the revenue that is gotten from that movie.

```python
# plot of popularity vs. profit
plt.figure(figsize=(12, 8))
plt.scatter(df['popularity'], df['profit'])
plt.title('Popularity vs Profit',fontsize = 15)
plt.xlabel('Popularity rating', fontsize = 14)
plt.ylabel('Profit', fontsize = 14)
```

There is a positive correlation between popularity and profit. The more popular the movie is, then the more the profit that it gives.
